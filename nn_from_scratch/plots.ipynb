```python
import matplotlib.pyplot as plt 
from math import exp
```

```python
def sigmoid_like(x):
    if x <0.4: 
        return 0
    if x >0.6: 
        return 1
    return 5*x-2

```

```python
def sigmoid(x: float): 
    # write the function of sigmoid
    sigmoid_output = exp(x)/(1 + exp(x))
    return sigmoid_output
```

```python
[sigmoid(0), sigmoid(0.3), sigmoid(0.3), sigmoid(0.6)]
```

```python
plt.plot([val for val in range(-10,11)], [sigmoid_like(x) for x in range(-10, 11)])
```
